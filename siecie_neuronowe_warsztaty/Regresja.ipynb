```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def forward(a,x):
    return a*x

def mse(y, yp):
    return np.mean((y-yp)**2)

def gradient(x,y,yp):
    return 2*np.mean((yp-y)*x)

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([1.1, 1.9, 3.2, 3.8, 5.2, 6.7])

a = 0
alpha = 0.05
a_print = []
L_print = []
for epoch in range(10):
    yp = forward(a,x)
    L = mse(y, yp)
    g = gradient(x,y,yp)
    a = a - g*alpha
    a_print.append(a)
    L_print.append(L)
```

```python
plt.plot(a_print)
plt.grid()
```

```python
plt.plot(L_print)
plt.grid()
```
